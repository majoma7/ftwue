```python
# n_pedestrians, n_pedestrians_towards, n_pedestrians_away as int

def convert_to_int(x: pd.Series):
    return x.astype(pd.Int64Dtype())

def create_full_date(x: pd.DataFrame):
    df = x.copy()
    df["date"] = df["date"].astype(str)
    df["hour"] = df["hour"].astype(str)
    df["full_date"] = df["date"] + " " + df["hour"]
    df["full_date"] = pd.to_datetime(df["full_date"])
    date_column = df["full_date"]
    return date_column

def clean_columns_data_all(df):
    df.drop(columns=["date", "hour", "city", "weekday"], inplace=True)
    return df

def clean_rows_data_all(df):
    df = df.drop(df[(df["full_date"] == "2019-10-27 02:00:00") & 
                    (df["collection_type"] == "no_information")].index)
    df = df.drop(df[(df["full_date"] == "2023-10-29 02:00:00") & 
                    (df["collection_type"] == "no_information")].index)
    return df

    

data_all["n_pedestrians"] = convert_to_int(data_all["n_pedestrians"])
data_all["n_pedestrians_towards"] = convert_to_int(data_all["n_pedestrians_towards"])
data_all["n_pedestrians_away"] = convert_to_int(data_all["n_pedestrians_away"])

# # combine data and hour to datetime
# data_all["date"] = data_all["date"].astype(str)
# data_all["hour"] = data_all["hour"].astype(str)
# data_all["full_date"] = data_all["date"] + " " + data_all["hour"]
# data_all["full_date"] = pd.to_datetime(data_all["full_date"])
data_all["full_date"] = create_full_date(data_all)

# data_all.drop(columns = ["date", "hour", "city", "weekday"], inplace=True)
# data_all = data_all.drop(data_all[(data_all["full_date"] == "2019-10-27 02:00:00") & 
#                                   (data_all["collection_type"] == "no_information")].index)
# data_all = data_all.drop(data_all[(data_all["full_date"] == "2023-10-29 02:00:00") & 
#                                   (data_all["collection_type"] == "no_information")].index)

data_all = clean_columns_data_all(data_all)
data_all = clean_rows_data_all(data_all)
```
